# Former i VPython (Glowscript)

Da vi ikke længere bruger Trinket, kører vi koden direkte i denne Notebook ved hjælp af `vpython` biblioteket. 

**Bemærk:** Hvis du kører dette lokalt, skal du først installere biblioteket med `pip install vpython`.

````python
from vpython import *

# Her er de forskellige former fra eksemplet:
sphere(pos=vec(-2,2,0), radius=0.5, color=color.red)
box(pos=vec(0,2,0), length=1, height=1, width=1, color=color.orange)
cylinder(pos=vec(2,2,0), axis=vec(0,1,0), radius=0.5, color=color.yellow)
cone(pos=vec(-2,0,0), axis=vec(0,1,0), radius=0.5, color=color.green)
pyramid(pos=vec(0,0,0), size=vec(1,1,1), color=color.cyan)
ring(pos=vec(2,0,0), axis=vec(0,0,1), radius=0.5, thickness=0.1, color=color.blue)
helix(pos=vec(-2,-2,0), axis=vec(1,0,0), radius=0.5, color=color.magenta)
````

### Øvelse: Forstå Akserne

For at få styr på akserne, kan du prøve at udkommentere (sætte `#` foran) linjerne i cellen ovenfor og i stedet indsætte disse tre pile:

```python
arrow(axis=vec(1,0,0), color=color.white)  # X-aksen
arrow(axis=vec(0,1,0), color=color.red)    # Y-aksen
arrow(axis=vec(0,0,1), color=color.blue)   # Z-aksen
```

* Vektoren er defineret som `vec(x,y,z)`.
